```python
top5_df=pd.read_csv('Top_5_categories.csv')
```

```python
top5_df
```

```python
month=top5_df['Month'].value_counts().reset_index()
month
```

```python
x=month['index'].sort_values()
y=month['Month']
label=['Jan','Feb','March','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
plt.figure(figsize=(8,5))
plt.bar(x,y)
plt.xticks(x,label)
for i, popu in enumerate(y):
    plt.text(i+1, popu, str(popu), ha='center',va='bottom')
plt.title('Month with most number of Posts')
plt.xlabel('Months')
plt.ylabel('Number of posts')
plt.show()
```

```python
top5_df['Category']=top5_df['Category'].str.lower()
```

```python
top5_df['Category']=top5_df['Category'].str.replace('"','')
```

```python
top5_df['Category'].value_counts()
```

```python
top5_df
```

```python
category_score = top5_df.groupby('Category')['Score'].sum()
```

```python
category_score_sorted = category_score.sort_values(ascending=False)
```

```python
category_score_sorted
```

```python
plt.figure(figsize=(25,8))
plt.bar(category_score_sorted.index,category_score_sorted.values)
plt.xlabel('Categories')
plt.ylabel('Scores')
plt.title('Top 5 Categories')
plt.show()
```

```python
top5_categories = ['animals','science','technology','travel','healthy eating']
```

```python
clean_df=top5_df.loc[top5_df['Category'].isin(top5_categories)]
```
